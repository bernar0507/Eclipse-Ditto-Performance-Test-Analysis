# Performance 1 iWatch vs 10 iWatches

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("performance10.csv")
df2 = pd.read_csv("performance1.csv")
```

```python
df1['Timestamp'] = pd.to_datetime(df1['Timestamp'])
df2['Timestamp'] = pd.to_datetime(df2['Timestamp'])

# Append the dataframes
df = df1.append(df2)
```

```python
# Set the timestamp as index
df.set_index('Timestamp', inplace=True)
```

```python
# Convert Time from seconds to minutes
df['Time'] = df['Time'] / 60
```

```python
plt.figure(figsize=(10,6))
for test in df['Test Name'].unique():
    plt.plot(df[df['Test Name'] == test]['Time'], df[df['Test Name'] == test]['CPU Usage (%)'], label=test)
plt.title('CPU Usage Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('CPU Usage (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot RAM usage over time for each test
plt.figure(figsize=(10,6))
for test in df['Test Name'].unique():
    plt.plot(df[df['Test Name'] == test]['Time'], df[df['Test Name'] == test]['RAM Usage (MB)'], label=test)
plt.title('RAM Usage Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('RAM Usage (MB)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Network usage over time for each test
plt.figure(figsize=(10,6))
for test in df['Test Name'].unique():
    plt.plot(df[df['Test Name'] == test]['Time'], df[df['Test Name'] == test]['Network Usage (Bytes)'], label=test)
plt.title('Network Usage Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Network Usage (Bytes)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Response time over time for each test
plt.figure(figsize=(10,6))
for test in df['Test Name'].unique():
    plt.plot(df[df['Test Name'] == test]['Time'], df[df['Test Name'] == test]['Response Time (s)'], label=test)
plt.title('Response Time Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Response Time (s)')
plt.legend()
plt.grid(True)
plt.show()
```
